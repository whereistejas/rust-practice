```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.misc import derivative
```

```python
xs = [x for x in np.arange(-1, 7, 0.0001)]
ys = np.zeros(len(xs))
yd = np.zeros(len(xs))
```

```python
def f(x):
    return (x-3)*(x**3)*(x-6)**4
```

```python
for (key, value) in enumerate(xs):
    x = value
    ys[key] = f(x)
    yd[key] = derivative(f, x, dx=1e-6)
    # ys[key] = (x - 6)**5 * (x**3*(x - 6)*(x - 3)**(x**4)*(x + 4*(x - 3)*math.log(x - 3)) + 6*(x - 3)**(x**4 + 1)) / (x - 3)
```

```python
plt.figure(figsize=(9,7), dpi=100)
plt.axhline(y=0, xmin=0, xmax=1, color='g')
plt.axvline(x=0, ymin=0, ymax=1, color='g')
plt.grid(True)
plt.plot(xs, ys)
#plt.plot(xs, yd)
plt.plot(0.5, f(0.5), 'ro')
#plt.plot(xs, yd)
```
