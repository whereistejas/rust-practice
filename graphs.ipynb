# rosenbrock

```python
def rosenbrock(X):
    """
    This R^2 -> R^1 function should be compatible with algopy.
    http://en.wikipedia.org/wiki/Rosenbrock_function
    A generalized implementation is available
    as the scipy.optimize.rosen function
    """
    x = X[0]
    y = X[1]
    a = 1. - x
    b = y - x*x
    return a*a + b*b*100.
```

```python
x = np.arange(-2, 2, 0.001)
y = np.arange(-2, 2, 0.001)
levels = np.arange(-2, 2, 0.1)

xx, yy = np.meshgrid(x, y, )
X = [xx, yy]
```

```python
result = rosenbrock(X)
```

```python
%matplotlib inline
fig = plt.figure(figsize=(9, 7), dpi=100)

ax = fig.gca(projection='3d')

plt.contour(x, y, result, 20)

plt.show()
```
